```python
import os
from dotenv import load_dotenv
from openai import OpenAI
```

```python
# importing Gradio
import gradio as gr
```

```python
load_dotenv(override = True)
openai_api_key = os.getenv('OPENAI_API_KEY')
anthropic_api_key = os.getenv('ANTHROPIC_API_KEY')
google_api_key = os.getenv('GOOGLE_API_KEY')
deepseek_api_key = os.getenv('DEEPSEEK_API_KEY')
groq_api_key = os.getenv('GROQ_API_KEY')
grok_api_key = os.getenv('GROK_API_KEY')


if openai_api_key:
    print(f"OpenAI API Key exists and begins {openai_api_key[:8]}")
else:
    print("OpenAI API Key not set")
    
if anthropic_api_key:
    print(f"Anthropic API Key exists and begins {anthropic_api_key[:7]}")
else:
    print("Anthropic API Key not set (and this is optional)")

if google_api_key:
    print(f"Google API Key exists and begins {google_api_key[:2]}")
else:
    print("Google API Key not set (and this is optional)")

if deepseek_api_key:
    print(f"DeepSeek API Key exists and begins {deepseek_api_key[:3]}")
else:
    print("DeepSeek API Key not set (and this is optional)")

if groq_api_key:
    print(f"Groq API Key exists and begins {groq_api_key[:4]}")
else:
    print("Groq API Key not set (and this is optional)")

if grok_api_key:
    print(f"Grok API Key exists and begins {grok_api_key[:4]}")
else:
    print("Grok API Key not set (and this is optional)")
```

```python
openai = OpenAI()

anthropic_url = "https://api.anthropic.com/v1/"
deepseek_url = "https://api.deepseek.com"
grok_url = "https://api.x.ai/v1"

anthropic = OpenAI(api_key = anthropic_api_key, base_url=anthropic_url)
deepseek = OpenAI(api_key=deepseek_api_key, base_url=deepseek_url)
grok = OpenAI(api_key=grok_api_key, base_url=grok_url)
```

```python
system_prompt = "You are an helful assistant, you help students to understand the cocnept clearly by simple explanation and examples"
```

```python
def stream_model(prompt, model_name):
    system_prompt = "You are an helful assistant, you help students to understand the cocnept clearly by simple explanation and examples"
    messages = [
        {"role":"system","content":system_prompt},
        {"role":"user", "content":prompt}
    ]
    model = ""
    if model_name == "GPT":
        response = openai.chat.completions.create(model = "gpt-4.1-mini", messages = messages, stream = True)
    elif model_name == "Claude":
        response = anthropic.chat.completions.create(model = "claude-sonnet-4-5-20250929", messages = messages, stream = True)
    elif model_name == "Deepseek":
        response = deepseek.chat.completions.create(model = "deepseek-reasoner", messages = messages, stream = True)
    elif model_name == "Grok":
        response = grok.chat.completions.create(model = "grok-4", messages = messages, stream = True)
    
    result = ""
    for chunk in response:
        result += chunk.choices[0].delta.content or ""
        yield result
    
```

```python
message_input = gr.Textbox(label = "Your Message", info = "Enter your message", lines= 4)
model_Selector = gr.Dropdown(["GPT", "Claude", "Deepseek","Grok"], label = "select Model", value = "GPT")
message_output = gr.Textbox(label="Response: ")

view = gr.Interface(
    fn = stream_model,
    title = "LLMs",
    inputs = [message_input, model_Selector],
    outputs = [message_output],
    examples = [
        ["Explain How Transforms Work??"],
        ["How Large Language Models understand Plain English?"]
    ],
    flagging_mode = "never"
)

view.launch()
```
