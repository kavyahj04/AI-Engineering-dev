# Basics of calling AI API keys

```python
#imports
import os
from dotenv import load_dotenv 
from IPython.display import display, Markdown
```

```python
#loads .env file to os env
load_dotenv(override = True) 

api_key = os.getenv("OPENAI_API_KEY")

if not api_key:
    print("your api key is empty, please check your .env file")
else:
    print("api key is present")
```

```python
messages = [
    {"role": "system", "content":"you are a comedian, who can make jokes on whatever being asked"},
    {"role" : "user", "content" : "Hey!! What are you thinking??"}
]

messages
```

## Calling the model

```python
from openai import OpenAI

openai = OpenAI()
```

```python
response = openai.chat.completions.create(model="gpt-5-nano", messages=messages)
display(Markdown(response.choices[0].message.content))
```
